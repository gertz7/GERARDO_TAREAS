```python

#Suma y grÃ¡fica de dos vectores

import matplotlib.pyplot as plt


def ingresar_vector(n):
    vector = [float(input(f"ingrese el valor {i+1}: "))
    for i in range(n)]
    return vector

def sumar_de_vectores(vector1, vector2):
    return [vector1 + vector2 for vector1, vector2 in zip (vector1, vector2)]

print ("ingrese el primer vector:")
vector1 = ingresar_vector(2)

print ("\ningrese el segundo vector:")
vector2 = ingresar_vector(2)

vector_resultado=sumar_de_vectores(vector1, vector2)

print ("\nPrimer vector  ingresado A: ", vector1)
print ("\nSegundo vector ingresado B: ", vector2)
print ("\nResultado de la suma de los dos vectores,  A + B :  ", vector_resultado)


# Grafica la suma
plt.quiver(0,0,vector1[0],  vector1[1],color='g',units='xy',scale=1)
plt.quiver(vector1[0],vector1[1],vector2[0],vector2[1],color='b',units='xy',scale=1)
plt.quiver(0,0,vector_resultado[0],vector_resultado[1],color='r',units='xy',scale=1)
plt.axis('scaled')

#Eje x
x1=vector1[0]
x2=vector2[0]

if x1>=0:
  if x2>=0:
    plt.xlim(0,1.1*(x1+x2))
    xa=0
    xb=1.1*(x1+x2)
  else:
    if x1+x2 >=0:
      plt.xlim(0,1.1*x1)
      xa=0
      xb=1.1*x1
    else:
      plt.xlim(1.1*(x1+x2),1.1*x1)
      xa=1.1*(x1+x2)
      xb=1.1*x1
else:
  if x2<=0:
    plt.xlim(1.1*(x1+x2),0)
    xa=1.1*(x1+x2)
    xb=0
  else:
    if x1+x2 <=0:
      plt.xlim(1.1*x1,0)
      xa=1.1*x1
      xb=0
    else:
      plt.xlim(1.1*x1,1.1*(x1+x2))
      xa=1.1*x1
      xb=1.1*(x1+x2)

#Eje y

y1=vector1[1]
y2=vector2[1]

if y1>=0:
  if y2>=0:
    plt.ylim(0,1.1*(y1+y2))
    ya=0
    yb=1.1*(y1+y2)
  else:
    if y1+y2 >=0:
      plt.ylim(0,1.1*y1)
      ya=0
      yb=1.1*y1
    else:
      plt.ylim(1.1*(y1+y2),1.1*y1)
      ya=1.1*(y1+y2)
      yb=1.1*y1
else:
  if y2<=0:
    plt.ylim(1.1*(y1+y2),0)
    ya=1.1*(y1+y2)
    yb=0
  else:
    if y1+y2 <=0:
      plt.ylim(1.1*y1,0)
      ya=1.1*y1
      yb=0
    else:
      plt.ylim(1.1*y1,1.1*(y1+y2))
      ya=1.1*y1
      yb=1.1*(y1+y2)

#Eje x-y
plt.hlines(y=0, xmin=xa, xmax=xb)
plt.vlines(x=0, ymin=ya, ymax=yb)
plt.show

```
